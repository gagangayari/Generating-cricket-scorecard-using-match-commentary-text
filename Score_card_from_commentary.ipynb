```python
## 

import re
##Function to limit a number upto 2 decimal places
def truncate(f, n):
    '''Truncates/pads a float f to n decimal places without rounding'''
    s = '{}'.format(f)
    if 'e' in s or 'E' in s:
        return '{0:.{1}f}'.format(f, n)
    i, p, d = s.partition('.')
    return '.'.join([i, (d+'0'*n)[:n]])



```

```python
## Function to calculate total_scores,extras,run_rate etc
def total_runs_extras(scores,curr_inngs):
    """
    Calculates TOTAL and the extra scores
    
    scores :like [[[]]] .3D list
    
    """
    total_1=0
    extras_1=0
    lb_1=0
    wide_1=0
    wicket_1 =0
    bye_1=0
    not_out_1=[]  ## list of not_out players
    over_1=curr_inngs[-1][1] 
    
    for index2 in scores:
        for index3 in index2:
            total_1 +=index3[2]
            lb_1 += index3[8]
            wide_1 += index3[9]
            bye_1 += index3[-1]
            if( index3[10]):
                wicket_1 +=1
            else: ## not_out player
                player_1=index3[0]
                not_out_1.append(player_1)
            
    extras_1=lb_1+wide_1
    
    if(wicket_1<11 and float(over_1)<50):
        over_1=50

    
    total_1 =total_1 +extras_1
    run_rate_1=total_1/over_1


    final_format="{} (b {},lb {},w {} )".format(extras_1,bye_1,lb_1,wide_1)
    total_format="{}/{} ({} overs,RR : {})".format(total_1,wicket_1,over_1,run_rate_1)
    
    print("\nExtras".ljust(70),final_format)
    print("TOTAL".ljust(69),total_format)
    
    return final_format

# generate_scorecard(bat_list,0)
```

```python
## Function to calculate fall_of_wickets
def fall_of_wickets(inngs_timeline):
    wide=re.compile("w$")
    leg_bye=re.compile("lb$")
    no_ball=re.compile("nb$")
    bye=re.compile('b$')
    total_run_1=0
    gone_wickets=0
    wicket_falls=[]
    
    for timeline in inngs_timeline:
        
        run_1 = timeline[0]
        if(run_1 =='W'):
            gone_wickets+=1
            out_at = timeline[1]
            batsman_1 = timeline[2]
            bowler_1 =timeline[3]
            
            output_format="{}-{}({}, {} ov)".format(gone_wickets,total_run_1,batsman_1,out_at)
            wicket_falls.append(output_format)
            
#             print(output_format)
            continue
        
            
            
        if(not re.search(leg_bye,run_1) and not re.search(wide,run_1) and not re.search(no_ball,run_1) and not re.search(bye,run_1)): ##if not wide,legbye or no_ball
                total_run_1 += int(run_1)

        else:
            if( re.search(leg_bye,run_1) or re.search(no_ball,run_1)): ## if "no_ball" or "leg_bye"
                total_run_1 += int(run_1[:-2])
            if( re.search(wide,run_1)):
                total_run_1 +=1
            if( re.search(bye,run_1) and not re.search(leg_bye,run_1)):
                total_run_1 += int(run_1[:-1])
    
    print("\nFall of wickets :  ",end='')
    for falls in wicket_falls:
        print(falls,end=' , ')
        

# generate_scorecard(bat_list)
```

```python
    
##* Batting squad timeline
def batsmans_timeline_(squad,team,timeline_number):
    if( re.match(team,squad) ):
        players=squad[len(team):].split(",")


    players_name=[]

    for player in players:
        digit=re.compile("[a-zA-Z]")
        k=(re.search(digit,player))
        i=player.index(k.group())
        players_name.append(player[i:])


    ## Calculating Batsman details
    bat_list=[]
    already_counted=[]
    for name in players_name:
        for name_with_details in both_inng_timeline[timeline_number]:
            name2=name_with_details[2]
            if(name.find(name2)!=-1 and name2 not in already_counted):
                batsman0=a_batsman(name2)  ## batsman object
                bat_list.append(batsman0.count_scores(both_inng_timeline[timeline_number]))
                already_counted.append(name2)
    return bat_list

```

```python
##* Bowling squad
def bowlers_timeline_(squad_,team_,timeline_):
    if( re.match(team_,squad_) ):
        players=squad_[len(team_):].split(",")


    players_name=[]
    for player in players:
    #     print(player)
        digit=re.compile("[a-zA-Z]")
        k=(re.search(digit,player))
        i=player.index(k.group())
        players_name.append(player[i:])

    ## Calculating Bowler details
    ball_list=[]
    already_counted=[]
    for name in players_name:
        for name_with_details in both_inng_timeline[timeline_]:
            name2=name_with_details[3]
            if(name.find(name2)!=-1 and name2 not in already_counted):
                bowler0=a_bowler(name2)
                ball_list.append(bowler0.count_bowler_scores(both_inng_timeline[timeline_]))
                already_counted.append(name2)
    return ball_list
```

```python

class a_batsman:
    def __init__(self,batsman1):
        self.name=batsman1
        self.run=0
        self.maiden=0
        self.six=0
        self.four=0
        self.balls=0
        self.lb=0
        self.w=0
        self.bye=0
        self.SR=0
        self.check_out_or_not=False  ## False = not out
        self.full_scorecard=[]
        self.bowler=None  ## bowler who took the batsman's wicket
#         print(self.batsman)
        
    
    def count_scores(self,scoreboard):
        wide=re.compile("w$")
        leg_bye=re.compile("lb$")
        bye=re.compile("b$")
        no_ball=re.compile("nb$")
        
        count_ball=0
        for index1,board in enumerate(scoreboard): 
            ## For maiden calculation
            count_ball+=1
            if(board[2]==self.name):
                
                run=board[0]
                if(run!='W' ):
                    if(not re.search(wide,run)):
                        self.balls+=1
                if(run=='W'):
                    self.balls+=1
                    self.check_out_or_not=True
                    self.bowler=board[3]
                    break
                if(run=='4'):
                    self.four+=1
                if(run=='6'):
                    self.six+=1
                if(re.search(bye,run) and not re.search(leg_bye,run)  ):
                    self.run+=int(run[:-1])
                    self.bye+=int(run[:-1])
                if(re.search(no_ball,run)):
                    self.run+=int(run[:-2])
                if(not re.search('lb$',run) and not re.search('w$',run) and not re.search(bye,run) and not re.search(no_ball,run)):#if not wide,legbye,noball or bye
                    self.run+=int(run)
                else:
                    if(re.search(leg_bye,run)):
                        self.lb+=int(run[:-2])
                        
                    if(re.search(wide,run)):
                        self.w+=1
                        
                ##For maiden calculation
                #calculate how long he played
                in_at_ball=index1+1
                if(run=='W'):
                    out_at_ball=count_ball
                try:
                    played_balls=out_at_ball-in_at_ball
                except:
                    ##If not out yet,thenn pass
                    flag1=1
                    pass
                if(flag1!=1):    
                    self.maiden=played_balls-self.run
                ##
                
#         self.balls=index1+1
        try:
            self.SR=truncate((self.run/self.balls)*100,2)
        except:
            self.SR="Did not bat"
            
        self.full_scorecard.append([self.name,self.bowler,self.run,self.balls,self.maiden,self.four,self.six,self.SR,self.lb,self.w,self.check_out_or_not,self.bye])
        
        ## extras
        extras=self.w + self.lb
        extras1="{} (lb {},w {})".format(extras,self.lb,self.w)
              
#         print("Extras : ",extras1)
        return self.full_scorecard
        
                
                
                
```

```python
class a_bowler:
    def __init__(self,bowler1):
        self.name=bowler1
        self.balls=0
        self.over=0
        self.run=0
        self.wicket=0
        self.fours=0
        self.six=0
        self.zeroes=0
        self.maiden=0
        self.economy=0
        self.no_balls=0
        self.wide=0
        self.bowlers_scorecard=[]
        
    def count_bowler_scores(self,scoreboard):
        wide=re.compile("w$")
        leg_bye=re.compile("lb$")
        no_ball=re.compile("nb$")
        bye=re.compile("b$")
        
        for balls in scoreboard:
            if(balls[3]==self.name):
                ball=balls[0]
                
                if(re.search(wide,ball)==None and re.search(no_ball,ball)==None):
#                     print(ball)
                    self.balls+=1
                    if(re.search(leg_bye,ball)):
                        self.run+=int(ball[:-2])
                    elif(re.search(bye,ball)):
                        self.run+=int(ball[:-1])
                    elif(ball!='W'):
                        self.run+=int(ball)
                    
                        
                if(re.search(wide,ball)):
                    self.run+=1
                    self.wide+=int(ball[:-1])
                if(re.search(no_ball,ball)):
                    self.no_balls+=1
                    self.run+=int(ball[:-2])
#                 if(re.search(leg_bye,ball) ):
#                     self.run+=int(ball[:-2])
#                 if(re.search(bye,ball) and not re.search(leg_bye,ball) and not re.search(no_ball,ball)):
#                     self.run+=int(ball[:-1])
                
                    
                if(ball=='6'):
                    self.six+=1
#                     self.run+=6
                if(ball=='4'):
#                     self.run+=4
                    self.fours+=1
                if(ball=='0'):
                    self.zeroes+=1
                if(ball=='W'):
                    self.wicket+=1
                    self.zeroes+=1
#         print(self.balls%6)
        
        if(self.balls<6):
            self.over=(0.1*self.balls)
        else:
            if(self.balls%6==0):
                self.over=self.balls/6
            else:
                quo=self.balls//6
                rem =self.balls%6
                self.over=quo+(0.1*rem)
        self.economy=truncate(self.run/self.over,2)
        
        self.bowlers_scorecard.append([self.name,self.over,self.maiden,self.run,self.wicket,self.economy,self.zeroes,self.fours,self.six,self.wide,self.no_balls])
        
        return self.bowlers_scorecard
```

```python
## Creating timeline of an inning in list format

both_inng_timeline=[]
for inng in innings:
    match_timeline=[]
    for j,line1 in enumerate((inng)):
        if(len(line1)<=4 and line1!='\n'):

            bowl_to_bats=inng[j-1].split(",")[0]         
            batsman=bowl_to_bats.split(" to ")[-1]
            bowler=bowl_to_bats.split(" to ")[0]
            balls=inng[j+1].replace("\n","")

            over=re.compile(r'\d\.\d')

            if(not re.match(over,line1)):
                run=line1.replace("\n","")
                match_timeline.append([run,balls,batsman,bowler])
    both_inng_timeline.append(match_timeline)

# both_inng_timeline[1]
```
